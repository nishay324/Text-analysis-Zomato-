# Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
url = "https://www.zomato.com/bangalore/voosh-thalis-bowls-1-bellandur-bangalore/order"

headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }

# send request to download the data
response = requests.request("GET", url, headers=headers)

# parse the downloaded data
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
cards_data =data.find_all('div', attrs= {'class':'sc-1s0saks-10 cYSFTJ'})
```

```python
# total number of cards
print('Total Number of Cards Found : ', len(cards_data))

```

```python
# source code of hotel cards
for card in cards_data:
    print(card)
```

```python
for card in cards_data:

    # get the Item name
    item_name = card.find('h4')
    # get the item price
    item_price = card.find('span', attrs={'class': 'sc-17hyc2s-1 cCiQWA'})
    # get description of the item
    item_description = card.find('p', attrs={'class':'sc-1s0saks-12 hcROsL'})
    
    print(item_name.text, item_price.text,item_description.text)
```

```python
# create a list to store the data
scraped_data = []

for card in cards_data:

    # initialize the dictionary
    card_details = {}

    # get the item name
    item_name = card.find('h4')

    # get the item price
    item_price = card.find('span', attrs={'class': 'sc-17hyc2s-1 cCiQWA'})
    
    # get the item description
    item_description = card.find('p', attrs={'class':'sc-1s0saks-12 hcROsL'})

    # add data to the dictionary
    card_details['item_name'] = item_name.text
    card_details['item_description'] = item_description.text
    card_details['item_price'] = item_price.text
    

    # append the scraped data to the list
    scraped_data.append(card_details)

# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(scraped_data)

# save the scraped data as CSV file
dataFrame.to_csv('restaurant_data.csv', index=False)
```

# Data Analysis ( Text Analysis)

```python
df= pd.read_csv('restaurant_data.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
# Data cleaning

# Removing Punctuation, numbers and special characters
df['item_clean_description']= df['item_description'].str.replace("[^a-zA-Z]"," ")
df.head(10)

# Removing short words
df['item_clean_description']=df['item_clean_description'].apply(lambda x : ' '.join(w for w in x.split() if len(w)>2))
    
```

**Text frequency**

TF = (Number of times term T appears in the particular row) / (number of terms in that row)

```python
tf1 = (df['item_clean_description'][1:]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
```

```python
tf1.head(50)
```

```python
tf1.tail(50)
```

**Inverse Document Frequency**

IDF = log(N/n)

```python
for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(df.shape[0]/(len(df[df['item_description'].str.contains(word)])))

tf1
```

**Term Frequency â€“ Inverse Document Frequency (TF-IDF)**

```python
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tfidf=tf1.sort_values(by=['tfidf'], axis=0, ascending=True, inplace=False) 
```

```python
top10=tfidf.head(10)
top10
```

```python
unique=tfidf.tail(10)
unique
```

```python
top10.to_csv('frequent_words.csv', index=False)
```

```python
unique.to_csv('unique_words.csv',index=False)
```
